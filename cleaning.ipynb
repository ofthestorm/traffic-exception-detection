```python
import numpy as np
import pandas as pd
from datetime import datetime
```

```python
f = open('./data/20111123/13301104001', 'r')        #以读方式打开文件
result = []
for line in f.readlines():                          #依次读取每行
    line = line.strip()                             #去掉每行头尾空白
    if not len(line) or line.startswith('#'):       #判断是否是空行或注释行
        continue                                    
    line = line.split(",")
    result.append(line)                             
# print(result)
```

```python
len(result)
```

```python
df = pd.DataFrame(result)
```

```python
df[3] =  pd.to_datetime(df[3], format='%Y%m%d%H%M%S')
df[4] = df[4].astype(float)
df[5] = df[5].astype(float)
df[8] = df[8].astype(int)
df[9] = df[9].astype(int)
```

```python
#五环内
df = df[df.apply(lambda x: x[4] > 116.2129590000 and x[4] < 116.5590580000, axis=1)]
df = df[df.apply(lambda x: x[5] > 39.7645780000 and x[5] < 40.0293980000, axis=1)]

#重复时间
df = df.drop_duplicates(3)

#速度90
df = df[df.apply(lambda x: x[8] <= 90, axis=1)]

#驻车
df = df[df.apply(lambda x: x[10] != "3", axis=1)]
```
