```python
import numpy as np
import pandas as pd
from sklearn.cluster import DBSCAN
from sklearn.cluster import KMeans
np.set_printoptions(suppress=True)
```

```python
csv_data=pd.read_csv('./temp_data/w.csv',header=None,sep=',')
df = pd.DataFrame(csv_data)
```

```python
# df.drop(df.index[0:120000], inplace=True)
# df
df
```

```python
db = DBSCAN(eps=0.1, min_samples=50).fit(df)
```

```python
pd.DataFrame(db.labels_).to_csv("./temp_data/db_result.txt", header=None, index=None)
```

```python
db
```

```python
kmeans = KMeans(n_clusters=200, random_state=0).fit(df)
```

```python
pd.DataFrame(kmeans.labels_).to_csv("./temp_data/km_result.txt", header=None, index=None)
```

```python
kmeans.predict([[0, 0, 1], [12, 3, 0]])
```
