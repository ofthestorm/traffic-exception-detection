```python
from sklearn.neighbors.kde import KernelDensity
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import pandas as pd
import scipy.stats as stats
import math
```

```python
# 参数定义
beta = 0.5
threshold = 0.1
k = 250
```

```python
#实时数据
realtime_csv_data=pd.read_csv('D:/dataminingclub/data2/20111124.csv',header=None,sep=',')
realtime_df_temp = pd.DataFrame(realtime_csv_data)
# realtime_df_temp.drop(0, axis=1, inplace=True)
# realtime_df_temp.drop(1, axis=1, inplace=True)
realtime_df_temp.drop(2, axis=1, inplace=True)
```

```python
#时段合并
for i in range(24):
    realtime_df_temp[99+i] = realtime_df_temp[4*i+3] + realtime_df_temp[4*i+4] + realtime_df_temp[4*i+5] + realtime_df_temp[4*i+6]

for i in range(3,99):
    realtime_df_temp.drop(i, axis=1, inplace=True)
```

```python
cluster_csv_data=pd.read_csv('D:/dataminingclub/kmeans_result.csv',header=None,sep=',')
cluster_df = pd.DataFrame(cluster_csv_data)
cluster_df = cluster_df.rename(columns={0:'cluster'})
```

```python
#加聚类标签
merged_realtime_df = pd.concat([realtime_df_temp,cluster_df],axis=1)
```

```python
grouped_realtime_dict = dict(tuple(merged_realtime_df.groupby('cluster')))

temp_realtime_list = []
cluster_realtime = {}  #实时数据

for i in range(k):
    temp_realtime_list = []
    for j in range(24):
        temp_realtime_list.append(grouped_realtime_dict[i][99+j].tolist())
    cluster_realtime[i] = temp_realtime_list
```

```python
#实时数据的均值和方差
realtime_u_sig={}
for i in range(250):
    realtime_1=[]
    for j in range(24):
        realtime_2=pd.DataFrame(cluster_realtime[i][j]).mean().values.tolist()
        realtime_2=realtime_2+pd.DataFrame(cluster_realtime[i][j]).var().values.tolist()
        realtime_1.append(realtime_2)
    realtime_u_sig[i]=realtime_1
```

```python
csv_data=pd.read_csv('D:/dataminingclub/wxh.csv',header=None,sep=',')
history_df = pd.DataFrame(csv_data)
```

```python
m_df = pd.concat([history_df,cluster_df],axis=1)
grouped_df = m_df.groupby('cluster')
```

```python
grouped_dict = dict(tuple(m_df.groupby('cluster')))

cluster_history={}
for i in range(len(grouped_dict)):
    temp_history_list = []
    for m in range(24):
        temp_history=[]
        for j in range(8):
            temp_history.append(grouped_dict[i][24*(j)+m])
        temp_history_list.append(temp_history)
    cluster_history[i]=temp_history_list
```

```python
#所有数据 格式√
history_u_sig={}
for i in range(250):
    history_2 = []
    for j in range(24):
        history_1=cluster_history[i][j][0].values.tolist()
        for m in range(7):
            history_1=history_1+cluster_history[i][j][m+1].values.tolist()
        history_2.append(history_1)
    history_u_sig[i]=history_2
```

```python
#得到了每一类在每一时刻的均值和方差
history_u_sig_1={}
for i in range(250):
    history_3=[]
    for j in range(24):
        history_4=pd.DataFrame(history_u_sig[i][j]).mean().values.tolist()
        history_4=history_4+pd.DataFrame(history_u_sig[i][j]).var().values.tolist()
        history_3.append(history_4)
    history_u_sig_1[i]=history_3
```

```python
merged_realtime_df[99+0][i]
```

```python

for i in range(len(merged_realtime_df)):  #遍历实时数据所有道路
    for j in range(24):    #遍历24个小时
        cluster_result=merged_realtime_df['cluster'][i]  #得到当前道路的聚类结果
        u_realtime=realtime_u_sig[cluster_result][j][0]   #得到当前类别和当前时间的均值
        sig_realtime=realtime_u_sig[cluster_result][j][1]   #得到当前类别和当前时间的方差
        Pr = stats.norm.cdf(merged_realtime_df[j+99][i],loc = u_realtime,scale = math.sqrt(sig_realtime))
        Pr=1-Pr   #得到横向比较的概率
    
        #历史
        u_history= history_u_sig_1[cluster_result][j][0] #得到历史当前类别和当前时间的均值
        sig_history=history_u_sig_1[cluster_result][j][1]  #得到历史当前类别和当前时间的方差
        Pa=stats.norm.cdf(merged_realtime_df[j+99][i],loc = u_history,scale = math.sqrt(sig_history))
        Pa=1-Pa  #得到纵向比较的概率
    
        #打分函数
        Score=beta*Pa+(1-beta)*Pr
        
        if Score<1e-9:
            print("路段代号：" ,end=" ")
            print(merged_realtime_df[0][i],end=' ')
            print(merged_realtime_df[1][i],end=' ')
            print(Score)
```
