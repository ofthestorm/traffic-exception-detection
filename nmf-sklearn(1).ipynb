```python
import numpy as np
import pandas as pd
from sklearn.decomposition import NMF
import matplotlib.pyplot as plt
import os
```

```python
#读取文件
#data=pd.read_csv("D:/dataminingclub/data2/20111116.csv",header=None)
os.listdir("D:/dataminingclub/data2")

csv_list=[]
for info in os.listdir("D:/dataminingclub/data2"):
    domain=os.path.abspath("D:/dataminingclub/data2")
    info=os.path.join(domain,info)
    csv_list.append(info)
print(csv_list)

data=[]
for i in csv_list:
    df=pd.read_csv(i,header=None)
    data.append(df)
```

```python
#删除前三列
for i in range(len(data)):
    data[i]=data[i].drop(0,axis=1)
    data[i]=data[i].drop(1,axis=1)
    data[i]=data[i].drop(2,axis=1)
```

```python
data[0]
```

```python
#sklearn nmf
#r为3
model=NMF(n_components=3,init='random',random_state=0)
W=[]
H=[]
for i in data:
    W.append(model.fit_transform(i))
    H.append(model.components_)
#W,H 为两个list 各有10个元素，为10个矩阵
```

```python
#10天数据求和取平均
W_sum=W[0]
for i in range(len(W)-1):
    W_sum=W_sum+W[i+1]

W_sum=W_sum/10
```
